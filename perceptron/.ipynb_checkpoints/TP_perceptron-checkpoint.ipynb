```python

def gen_time (sigma,M,n):
    
    Xa1 = np.random.uniform(0,M,n)
    Xb1 = np.zeros(n)
    X1 = np.concatenate((Xa1.reshape(-1,1),Xb1.reshape(-1,1)),axis=1)

    Xa2 = np.random.uniform(0,M,n)
    Xb2 = np.repeat(sigma,n)
    X2 = np.concatenate((Xa2.reshape(-1,1),Xb2.reshape(-1,1)),axis=1)

    X = np.concatenate((X1,X2))

    y = np.hstack([np.ones(n), -1 * np.ones(n)])
    return(X,y)
```

```python
import time
t0 = time.time()
gs_time = gradient(gen_time (1,2,100)[0], gen_time (1,2,100)[1], eps = 0.1, niter = 1000, w_ini = np.array([1,1,1])
         , loss_fun=hinge_loss, gr_loss_fun = gr_hinge_loss, stochastic=True)
print(time.time() - t0)
```

9) Proposez/codez des variantes de conditions d'arrêt pour l'algorithme.

```python
def gradient_arret_gradient(x, y, eps, alpha_stop, w_ini, loss_fun, gr_loss_fun, stochastic=True):
    """ algorithme de descente du gradient:
        - x : donnees
        - y : label
        - eps : facteur multiplicatif de descente
        - niter : nombre d'iterations
        - w_ini
        - loss_fun : fonction de cout
        - gr_loss_fun : gradient de la fonction de cout
        - stoch : True : gradient stochastique
        """
    w = []
    w.append(w_ini)
    loss = []
    loss.append(loss_fun(x, y, w[0]).mean())
    alpha = np.dot(gr_loss_fun(x,y, w_ini),np.transpose(gr_loss_fun(x,y, w_ini)))
    i=0
    while alpha > alpha_stop:
        i+=1
        if stochastic:  # this is for Stochastic Gradient Descent
            idx = [np.random.randint(x.shape[0])]
        else:           # this is for pure Gradient Descent
            idx = np.arange(x.shape[0])
        a=w[i-1]
        w.append(w[i - 1] - eps * gr_loss_fun(x[idx, :],y[idx], w[i - 1])) 
        alpha = np.dot(gr_loss_fun(x[idx, :], y[idx], w[i]),np.transpose(gr_loss_fun(x[idx, :],y[idx], w[i])))
        loss.append(loss_fun(x, y, w[i]).mean())
    return w, loss
```

```python
def gradient_arret_un(x, y, eps, beta, w_ini, loss_fun, gr_loss_fun, stochastic=True):
    """ algorithme de descente du gradient:
        - x : donnees
        - y : label
        - eps : facteur multiplicatif de descente
        - niter : nombre d'iterations
        - w_ini
        - loss_fun : fonction de cout
        - gr_loss_fun : gradient de la fonction de cout
        - stoch : True : gradient stochastique
        """
    w = []
    w.append(w_ini)
    loss = []
    loss.append(loss_fun(x, y, w[0]).mean())
    alpha = np.dot(gr_loss_fun(x,y, w_ini),np.transpose(gr_loss_fun(x,y, w_ini)))
    i=0
    while alpha > beta:
        i+=1
        if stochastic:  # this is for Stochastic Gradient Descent
            idx = [np.random.randint(x.shape[0])]
        else:           # this is for pure Gradient Descent
            idx = np.arange(x.shape[0])
        a=w[i-1]
        w.append(w[i - 1] - eps * gr_loss_fun(x[idx, :],y[idx], w[i - 1]))
        loss.append(loss_fun(x, y, w[i]).mean())
        alpha = np.dot(w[i]-w[i-1],w[i]-w[i-1])**2
        
    return w, loss
```

```python
def plot_gradient_custom(X, y, wh, cost_hist, loss_fun):
    """ display 4 figures on how  (stochastic) gradient descent behaves
    wh : solution history
    cost_hist : cost history
    loss_fun : loss function
    """
    best = len(cost_hist)-1
    plt.subplot(221)
    plt.title('Data and hyperplane estimated')
    plot_2d(X, y, wh[best])
    plt.subplot(222)
    plt.title('Projection of level line and algorithm path')
    plot_cout(X, y, loss_fun, wh)
    plt.subplot(223)
    plt.title('Objective function vs iterations')
    plt.plot(range(cost_hist.shape[0]), cost_hist)
    plt.subplot(224, projection='3d')
    plt.title('Level line and algorithm path')
    plot_cout3d(X, y, loss_fun, wh)
```

```python
def gradient_stop(x, y, eps, beta, w_ini, loss_fun, gr_loss_fun, stochastic=True):
    """ algorithme de descente du gradient:
        - x : donnees
        - y : label
        - eps : facteur multiplicatif de descente
        - niter : nombre d'iterations
        - w_ini
        - loss_fun : fonction de cout
        - gr_loss_fun : gradient de la fonction de cout
        - stoch : True : gradient stochastique
        """
    w = np.zeros((niter, w_ini.size))
    w[0] = w_ini
    loss = np.zeros(niter)
    loss[0] = loss_fun(x, y, w[0]).mean()
    for i in range(1, niter):
        if stochastic:  # this is for Stochastic Gradient Descent
            idx = [np.random.randint(x.shape[0])]
        else:           # this is for pure Gradient Descent
            idx = np.arange(x.shape[0])
        w[i, :] = w[i - 1, :] - eps * gr_loss_fun(x[idx, :],
                                                  y[idx], w[i - 1, :])
        loss[i] = loss_fun(x, y, w[i, :]).mean()
    return w, loss
```
